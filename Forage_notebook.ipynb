```python
# Get a copy of the data
new_df = df.copy()
```

```python
# Check duplicate rows
new_df[new_df.duplicated()]
```

```python
# Drop duplicated rows
new_df.drop_duplicates(inplace=True, ignore_index=True)
```

```python
new_df.head()
```

```python
new_df.duplicated().sum()
```

```python
# Drop rows where Unitprice is less than 0 and Quantity is less than 1
new_df = new_df.query('UnitPrice >= 0')
new_df = new_df.query('Quantity >= 1')
```

```python
# Check if rows with UnitPrice less than 0 are dropped
new_df[new_df.UnitPrice <0]
```

```python
# Check if rows with Quantity less than 1 are dropped
new_df[new_df.Quantity < 0]
```

```python
new_df.info()
```

```python
# Create a revenue column
new_df['Revenue'] = new_df.UnitPrice * new_df.Quantity
new_df.head()
```

```python
# Save cleaned data as Excel file
new_df.to_excel('online_retailsss.xlsx', index=False)
```
